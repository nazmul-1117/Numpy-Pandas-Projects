# 7.Perform Data Cleaning [ Replace '?' with NaN ]

````python
nd = df.copy()
print(nd.isin(['?']).sum())
print()

nd['workclass'] = nd['workclass'].replace('?', np.nan)
nd['occupation'] = nd['occupation'].replace('?', np.nan)
nd['native-country'] = nd['native-country'].replace('?', np.nan)

print(nd.isin(['?']).sum())
````

````python
nd.isnull().sum()
````

````python
nd.isnull().sum()
````

````python
nd.shape
````

# 8.1 Before drop see the percentage

````python
x = (nd.isnull().sum()/nd.shape[0])*100
x
````

````python
nd = nd.dropna(how='any')
nd.isnull().sum()
````

````python
nd.shape
````

# 9. Check for Duplicate Data and Drop Them

````python
dp = nd.copy()
print(dp.shape)
print()


d = dp.duplicated().any()
dp.drop_duplicates(inplace=True)

print(f'Missing Value: {d}')
print(dp.shape)
````
