# 8. How many people have French 'fr' as their Language?

# 11. How many People have Mastercard as their Credit Card Provider and made a purchase above 50?

````python
data.head(2)
x = data[data['CC Provider'] == 'Mastercard']
x = x[x['Purchase Price'] > 50]
print(x['Credit Card'].count())


# another way
print(data[(data['CC Provider'] == 'Mastercard') & (data['Purchase Price'] > 50)].count())
````

# 13. How many people purchase during the AM and how many people purchase during PM?

````python
data.head(2)
x = data['AM or PM']
x = x[x == 'AM']
print(x.count())

y = data[data['AM or PM'] == 'PM']
print(y['AM or PM'].count())
print()

# another way
print(data['AM or PM'].value_counts())
````

# 14. How many people have a credit card that expires in 2020?

````python
# first method using user define function for count value

def sum():
  count = 0
  for date in data['CC Exp Date']:
    if date.split('/')[1] == '20':
      count = count + 1
  print(count)

sum()
````

````python
# Another way to solve this using lambda function
x = data['CC Exp Date']
x = x.apply(lambda x: x[3:] == '20')
print(x.sum())
````
