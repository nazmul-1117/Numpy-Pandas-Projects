#17.    Display Name of the Person Having the Highest BasePay

````python
x = data.copy()

x['BasePay'] = x['BasePay'].replace('Not Provided', np.nan)
x['BasePay'] = x['BasePay'].fillna(value=0)
x['BasePay'] = x['BasePay'].astype(float)

x.head()
````

````python
y = x['BasePay'].max()
print(f'Highest BasePay: {y}')

y = x[x['BasePay'] == y]
print(y[['EmployeeName', 'JobTitle','BasePay', 'TotalPayBenefits']])
````
